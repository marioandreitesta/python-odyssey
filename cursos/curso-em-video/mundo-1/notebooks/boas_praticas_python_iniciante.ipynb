## 1. O Exercício #015 Refatorado

Aqui está o código original refatorado com as melhores práticas:

```python
# --- CONFIGURAÇÕES (CONSTANTES) ---
# Usamos letras MAIÚSCULAS para indicar valores que não mudam (constantes)
PRECO_DIA = 60.00
PRECO_KM = 0.15

print("="*40)
print(f"{'LOCADORA DE VEÍCULOS':^40}")
print("="*40)

# --- ENTRADA DE DADOS ---
try:
    dias = int(input("Quantidade de dias de aluguel: "))
    km = float(input("Quantidade de Km percorridos:  "))
except ValueError:
    print("ERRO: Digite apenas números válidos!")
    dias = km = 0

# --- CÁLCULOS ---
pago_dias = dias * PRECO_DIA
pago_km = km * PRECO_KM
total = pago_dias + pago_km

# --- SAÍDA DE DADOS (USANDO F-STRINGS) ---
print("-" * 40)
print(f"Resumo do Aluguel:")
print(f"-> {dias} dias (R${PRECO_DIA:.2f}/dia): R${pago_dias:>10.2f}")
print(f"-> {km:.1f} Km (R${PRECO_KM:.2f}/Km):  R${pago_km:>10.2f}")
print("-" * 40)
print(f"TOTAL A PAGAR: {'R$':>17}{total:.2f}")
print("=" * 40)
```

## 2. Pilares das Boas Práticas

### A. Fuja dos "Números Mágicos"
Um "número mágico" é um valor solto no meio do código (ex: `60.0`). Se você precisar mudar o preço do aluguel, terá que caçar esse número em todo o programa.

**Dica:** Defina **Constantes** (em letras maiúsculas) no topo do arquivo.

### B. Nomenclatura de Variáveis (`snake_case`)
No Python, o padrão para nomes de variáveis é o `snake_case` (letras minúsculas separadas por underline).

*   ❌ `D` ou `PrecoDoCarro` (CamelCase é para classes)
*   ✅ `dias_alugados` ou `preco_total` 

### C. Alinhamento com f-strings
As f-strings são muito poderosas para criar relatórios alinhados:

*   `{valor:.2f}`: Formata com 2 casas decimais.
*   `{texto:^20}`: Centraliza o texto em 20 espaços.
*   `{valor:>10.2f}`: Alinha o número à direita (perfeito para preços).

### D. Segredos do Alinhamento (`:>10.2f` e `:>17`)

Você notou códigos estranhos como `:>10.2f` no exercício? Eles são a chave para um visual profissional.

**O que cada símbolo significa:**
*   `:` → Indica o início da formatação.
*   `>` → Alinha o conteúdo à **Direita**.
*   `<` → Alinha o conteúdo à **Esquerda**.
*   `^` → **Centraliza** o conteúdo.
*   `10` ou `17` → O **Tamanho total** do espaço reservado (em caracteres).
*   `.2f` → Exibe exatamente **2 casas decimais** (f de *float*).

#### Exemplo Comparativo:

```python
item1, preco1 = "Pão", 1.50
item2, preco2 = "Leite Integral", 5.80

print("SEM ALINHAMENTO:")
print(f"{item1}: R${preco1}")
print(f"{item2}: R${preco2}")

print("\nCOM ALINHAMENTO (:.<20 e :>8.2f):")
print(f"{item1:.<20} R${preco1:>8.2f}")
print(f"{item2:.<20} R${preco2:>8.2f}")

# O ponto (.) antes do < ou > serve para preencher o espaço vazio com pontos!
```

## 3. Próximos Passos: O Padrão PEP 8

O **PEP 8** é o guia de estilo oficial do Python. Aqui estão 3 regras de ouro:

1. **Espaçamento:** Sempre use 1 espaço em volta de operadores (ex: `a = b + c`).
2. **Linhas em branco:** Deixe linhas em branco para separar blocos de código (ex: entrada de dados de cálculos).
3. **Comentários:** Comente **POR QUE** você fez algo, não **O QUE** o código faz (se o código for claro, o 'o que' já é óbvio).

```python
# Exemplo de código "sujo" vs "limpo"

# RUIM
x=10
y=2
print(x*y)

# BOM
quantidade_itens = 10
preco_unitario = 2.00

valor_venda = quantidade_itens * preco_unitario
print(f"O valor da venda é R${valor_venda:.2f}")
```
